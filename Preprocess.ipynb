## 1. Dummy Mapping

1. **AKD** likely stands for *Alat Kelengkapan Dewan*, which refers to internal bodies or committees within the Indonesian House of Representatives (DPR), such as Komisi Iâ€“XI, Baleg, BURT, and others.

2. **Sifat rapat** most likely indicates the type or privacy level of a meeting, with possible values including:
   - **Terbuka** â€” Open to public
   - **Tertutup** â€” Closed meeting
   - **Rahasia** â€” Confidential or sensitive meeting
   - **Tidak Diketahui** â€” Unknown or unrecorded status

3. **Jenis agenda** is currently undefined in the dataset. However, for dummy data purposes, it is logical to assume agenda types such as:
   - Rapat Kerja
   - Rapat Dengar Pendapat
   - Kunjungan Kerja
   - Pembahasan RUU
   - Workshop


```python
akd_mapping = {
    1: "Komisi I", 2: "Komisi II", 3: "Komisi III", 4: "Komisi IV", 5: "Komisi V",
    6: "Komisi VI", 7: "Komisi VII", 8: "Komisi VIII", 9: "Komisi IX", 10: "Komisi X",
    11: "Komisi XI", 12: "Badan Legislasi", 13: "Badan Anggaran", 14: "Badan Kehormatan",
    15: "BURT", 16: "BKSAP"
}
sifat_mapping = {1: "Terbuka", 2: "Tertutup", 3: "Rahasia", 0: "Tidak diketahui"}
status_mapping = {1: "Terjadwal", 9: "Dibatalkan"}

jenis_agenda_map = {
    1: "Rapat Kerja",               # paling umum
    2: "Rapat Dengar Pendapat",     # paling banyak muncul
    3: "Kunjungan Kerja",
    4: "Rapat Internal",
    5: "Rapat Gabungan",
    6: "Seminar / Diskusi",
    7: "Sosialisasi",
    9: "Public Hearing",
    11: "Evaluasi Program",
    12: "Rapat Penyusunan",
    15: "Rapat Koordinasi",
    16: "Rapat Tertutup",
    18: "Kegiatan Eksternal",
    22: "Workshop / Bimtek",
    26: "Uji Kepatutan dan Kelayakan",
    27: "Peninjauan Lapangan",
    30: "Rapat Anggaran",
    31: "Rapat Panja",
    34: "Studi Banding"
}


df['id_akd'] = df['id_akd'].map(akd_mapping)
df['sifat_rapat'] = df['sifat_rapat'].map(sifat_mapping)
df['status'] = df['status'].map(status_mapping)
df['date'] = df['date'].fillna("Tidak diketahui")
df['id_jenis_agenda'] = df['id_jenis_agenda'].map(jenis_agenda_map).fillna("tidak diketahui")
```
